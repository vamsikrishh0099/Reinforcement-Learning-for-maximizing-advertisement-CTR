```python
df=pd.read_csv('Ads_CTR_Optimisation.csv')
#this is a simulation of user choice 


#rows=customers
#columns=ads which a corresponding user would click if they appear on the screen
# example:the user in row 1 would click on the ad only if ad 9 appears on the screen.
#         the user in row 3 would click on the ad only if ad 2 or ad 8 appears on the screen
```

```python
#RANDOM SELECTION ALGORITHM

```

```python
import random
```

```python
N=10000
d=10
ads_selected=[]
total_reward=0
for n in range(0,N):
    ad=random.randrange(d)
    ads_selected.append(ad)
    reward=df.values[n,ad]
    total_reward=total_reward+reward
    
```

```python
#UPPER CONFIDENCE BOUND ALGORITHM

```

```python
number_of_selections=[0]*10
```

```python
sum_of_rewards=[0]*10
ads_selected=[]
```

```python
N=10000
d=10
import math
total_reward=0
```

```python
for n in range(0,N):
    max_upper_bound=0
    ad=0
    for i in range(0,d):
        if number_of_selections[i]>0:
            
            avg_reward=sum_of_rewards[i]/number_of_selections[i]     
            delta_i=math.sqrt(3/2*(math.log(n+1)/number_of_selections[i]))  #confidence interval 
            upper_bound=avg_reward+delta_i                                  #upper confidence bound
        else:
            upper_bound=1e400
        if upper_bound>max_upper_bound:
            max_upper_bound=upper_bound
            ad = i
    ads_selected.append(ad)
    number_of_selections[ad]+=1 
    reward=df.iloc[n,ad]
    sum_of_rewards[ad]+=reward
    total_reward=total_reward+reward
```
